```python
import pandas as pd
import json
```

```python
file_path= r'G:\NLP Projects\MitRA aTTACK\Notre projet\Data\tactic.json'
with open(file_path, 'r') as file:
    data = json.load(file)

data[1]
```

```python
records = []

for tactic in data:
    tactic_name = tactic.get('tactic-name')
    tactic_id = tactic.get('ID')
    tactic_description = tactic.get('tactic')
    
    for technique_group in tactic.get('technique', []):
        for technique in technique_group:
            technique_name = technique.get('technique-name')
            technique_id = technique.get('technique-ID')
            technique_description = technique.get('technique-description')
            
            for sub_technique_group in technique.get('sub-technique', []):
                for sub_technique in sub_technique_group:
                    sub_technique_name = sub_technique.get('sub-technique-name')
                    sub_technique_number = sub_technique.get('number')
                    sub_technique_description = sub_technique.get('sub-technique-description')
                    
                    records.append({
                        'technique_id': technique_id,
                        'sub_technique_name': sub_technique_name,
                        'sub_technique_number': sub_technique_number,
                        'sub_technique_description': sub_technique_description
                    })
                    
df_subtechnique = pd.DataFrame(records)

df_subtechnique
```

```python
tactics_list = []
for tactic in data:
    tactic_name = tactic.get('tactic-name')
    tactic_id = tactic.get('ID')
    tactic_description = tactic.get('tactic')
    tactics_list.append({'tactic_name': tactic_name, 'tactic_id': tactic_id, 'tactic_description': tactic_description})

df_tactic = pd.DataFrame(tactics_list)

df_tactic
```

```python
technique_list = []
for tactic in data:
    tactic_name = tactic.get('tactic-name')
    tactic_id = tactic.get('ID')
    tactic_description = tactic.get('tactic')
    
    for technique_group in tactic.get('technique', []):
        for technique in technique_group:
            technique_name = technique.get('technique-name')
            technique_id = technique.get('technique-ID')
            technique_description = technique.get('technique-description')
            technique_list.append({
                'tactic_id': tactic_id,
                'technique_name': technique_name,
                'technique_id': technique_id,
                'technique_description': technique_description
            })

df_technique = pd.DataFrame(technique_list)

df_technique
```

```python
def concatenate_strings(string_list):
    return '\n'.join(string_list)
```

```python
df_subtechnique['sub_technique_description'] = df_subtechnique['sub_technique_description'].apply(concatenate_strings)
df_tactic['tactic_description'] = df_tactic['tactic_description'].apply(concatenate_strings)
df_technique['technique_description'] = df_technique['technique_description'].apply(concatenate_strings)
```

```python
df_subtechnique
```

```python
def merge_technique_and_sub(technique_id, sub_technique_number):
    return f"{technique_id}.{sub_technique_number}"

df_subtechnique['sub_technique_number'] = df_subtechnique.apply(lambda row: merge_technique_and_sub(row['technique_id'], row['sub_technique_number']), axis=1)

df_subtechnique
```
